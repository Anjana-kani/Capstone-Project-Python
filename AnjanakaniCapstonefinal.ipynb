```python
#Task 2
#The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values.
#Replace the missing values by running average. You should use the “For” loop for this task.
```

```python
#Importing the project file as dataframe

project = pd.read_csv("D:\SkilloVilla\project_data.csv", index_col=0)
project
```

```python
#Creating a list of running averages using 'Cost' column in the project dataframe

running_avg = pd.Series(project['Cost'].rolling(window=3, center=True, min_periods=1).mean())
running_avg = running_avg.astype(float).round()
running_avg
```

```python
#inserting the sRunning Average Series as a column in the dataframe beside the 'Cost' column.
project.insert(3,"RunAvgCost", running_avg)
```

```python
#Replace the missing values by running average using the “For” loop and droping the avg.

df = pd.DataFrame(project)
for i in range(len(df)):
    if pd.isna(df.loc[i, 'Cost']):
        df.loc[i, 'Cost'] = df.loc[i, 'RunAvgCost']

df.drop('RunAvgCost',inplace = True, axis=1)
project_new = df
project_new
```

```python
# saving into new file
project_new.to_csv('D:\SkilloVilla\project_new_updated.csv', index=0)
```

```python
#Task 5
#Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects.
```

```python
#adding bonus 

list_bonus=[]

for i, r in final.iterrows():
    if(final['Status'].iat[i] == 'Finished'):
        result=(r['Cost']) + (r['Cost'])*0.05
        list_bonus.append(result)
    else:
        result=(r['Cost']) + (r['Cost'])*0
        list_bonus.append(result)

final.insert(8, "Bonus", list_bonus)
```

```python
#Task 6
#Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4.
```

```python

new_des = []

for i, r in final.iterrows():
    if(final["Status"].iat[i] == 'Failed'):
        result=(r['Designation Level']) - 1
        new_des.append(result)
    else:
        result=(r['Designation Level'])-0
        new_des.append(result)

final.insert(5, 'Designation', new_des)
```

```python
#Task 8
#Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.
```

```python
#creating new column with desigantion level;
new_des = []

for i, r in final.iterrows():
    if(final["Age"].iat[i] > 29):
        result=(r['Designation']) + 1
        new_des.append(result)
    else:
        result=(r['Designation'])+0
        new_des.append(result)

final.insert(6, 'Designation Level', new_des)
```
